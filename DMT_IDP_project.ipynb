```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler  # Import MinMaxScaler for normalization

# Load the dataset
data = pd.read_csv('/content/autism.csv')  # Replace with your actual dataset path

# Display the first few rows of the dataset to understand its structure
print("Original Dataset:")
print(data.head())

# Step 1: Handling Missing Values
# Replace missing values in categorical columns with 'missing'
categorical_columns = data.select_dtypes(include=['object']).columns
for column in categorical_columns:
    data[column] = data[column].fillna('missing')

# Replace missing values in numerical columns with the median
numerical_columns = data.select_dtypes(include=['number']).columns
for column in numerical_columns:
    data[column] = data[column].fillna(data[column].median())

print("\nDataset after handling missing values:")
print(data.head())

# Step 2: Encoding Categorical Variables
# Label Encoding for binary and nominal categorical variables
label_encoder = LabelEncoder()
for column in categorical_columns:
    # Convert all categorical values to string to ensure consistency
    data[column] = data[column].astype(str)
    # Apply label encoding to each categorical column
    data[column] = label_encoder.fit_transform(data[column])

print("\nDataset after encoding categorical variables:")
print(data.head())

# Step 3: Feature Normalization
# Normalizing numerical features to the range [0, 1]
scaler = MinMaxScaler()
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])

print("\nDataset after feature normalization:")
print(data.head())

# Save the preprocessed data for further use
data.to_csv('data_preprocessed_normalized.csv', index=False)
print("\nPreprocessed data saved as 'data_preprocessed_normalized.csv'.")
```

Federated learning



```python
import pandas as pd
import numpy as np

# Load the preprocessed dataset
data = pd.read_csv('/content/data_preprocessed_normalized.csv')  # Use your preprocessed data file

# Ensure the target column is retained
target_column = 'Class/ASD Traits '
if target_column not in data.columns:
    raise ValueError(f"Target column '{target_column}' not found in the dataset.")

# Simulate federated learning setup by splitting data into multiple nodes
num_nodes = 3  # Number of simulated nodes
data_splits = np.array_split(data, num_nodes)

# Display sizes of each node's dataset to verify distribution
for i, node_data in enumerate(data_splits):
    print(f"Node {i+1} dataset size: {node_data.shape[0]}")
```

```python
from sklearn.linear_model import LogisticRegression

# Train a logistic regression model on each node's data
local_models = []
for i, node_data in enumerate(data_splits):
    X_local = node_data.drop(columns=[target_column])
    y_local = node_data[target_column]

    # Train the local model
    local_model = LogisticRegression(random_state=42, max_iter=1000)
    local_model.fit(X_local, y_local)

    # Store the trained model
    local_models.append(local_model)
    print(f"Local Model {i+1} training complete.")
```

```python
# Function to aggregate local model weights
def aggregate_weights(local_models):
    global_weights = np.mean([model.coef_ for model in local_models], axis=0)
    global_intercept = np.mean([model.intercept_ for model in local_models], axis=0)
    return global_weights, global_intercept

# Aggregate weights to create a global model
global_weights, global_intercept = aggregate_weights(local_models)
global_model = LogisticRegression(random_state=42, max_iter=1000)
global_model.coef_ = global_weights
global_model.intercept_ = global_intercept
print("Global model created from local models.")
```

```python
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the preprocessed data (the dataset on which the global model is based)
data = pd.read_csv('/content/data_preprocessed_normalized.csv')  # Use your preprocessed data file

# Separate features and target
target_column = 'Class/ASD Traits '
X = data.drop(columns=[target_column])
y = data[target_column]

# Step 1: Apply PCA and determine the optimal number of components
pca_full = PCA()
pca_full.fit(X)
explained_variance = pca_full.explained_variance_ratio_
cumulative_explained_variance = explained_variance.cumsum()

# Choose number of components to capture at least 90% of variance
desired_variance = 0.90
n_components = next(i for i, cumulative_var in enumerate(cumulative_explained_variance) if cumulative_var >= desired_variance) + 1

# Print the number of components chosen
print(f"Number of components chosen: {n_components} out of {X.shape[1]} components.")

# Step 2: Apply PCA with the optimal number of components
pca_optimal = PCA(n_components=n_components)
X_pca = pca_optimal.fit_transform(X)

# Create a DataFrame with PCA-transformed data and retain the target column
pca_columns = [f'PC{i+1}' for i in range(n_components)]
pca_df = pd.DataFrame(data=X_pca, columns=pca_columns)
pca_df[target_column] = y.values

# Display the head of the PCA-transformed data
print("\nHead of the PCA-transformed data:")
print(pca_df.head())

# Save the PCA-transformed data to a new CSV file
pca_file_path = 'pca_transformed_autism_data.csv'
pca_df.to_csv(pca_file_path, index=False)
print(f"\nPCA-transformed data saved to {pca_file_path}")

# Step 3: Graphical Representation of Cumulative Explained Variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='--')
plt.axhline(y=desired_variance, color='r', linestyle='--', label=f'{desired_variance*100}% Variance Threshold')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance vs. Number of Principal Components')
plt.xticks(range(1, len(cumulative_explained_variance) + 1))
plt.grid(True)
plt.legend()
plt.show()
```

Train test split


```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load the PCA-transformed data
pca_file_path = 'pca_transformed_autism_data.csv'  # Path to the saved PCA-transformed data file
pca_df = pd.read_csv(pca_file_path)

# Display the head of the PCA-transformed data to confirm loading
print("Loaded PCA-transformed data:")
print(pca_df.head())

# Separate features and target from the PCA-transformed data
target_column = 'Class/ASD Traits '
X_pca = pca_df.drop(columns=[target_column])
y_pca = pca_df[target_column]

# Perform train-test split
# Use 80% of the data for training and 20% for testing
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y_pca, test_size=0.2, random_state=42)

# Display the shapes of the train and test sets to confirm the split
print("\nTrain-Test Split Results:")
print(f"Training set (X_train_pca): {X_train_pca.shape}, Training target (y_train_pca): {y_train_pca.shape}")
print(f"Testing set (X_test_pca): {X_test_pca.shape}, Testing target (y_test_pca): {y_test_pca.shape}")
```

Applying Machine Learining Alogrithms


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt

# Train Logistic Regression on PCA-transformed training data
log_reg_model = LogisticRegression(random_state=42, max_iter=1000)
log_reg_model.fit(X_train_pca, y_train_pca)

# Evaluate the model
y_pred_log_reg = log_reg_model.predict(X_test_pca)
accuracy_log_reg = accuracy_score(y_test_pca, y_pred_log_reg)

# Calculate precision, recall, and f1 score
precision_log_reg = precision_score(y_test_pca, y_pred_log_reg, average='binary')
recall_log_reg = recall_score(y_test_pca, y_pred_log_reg, average='binary')
f1_log_reg = f1_score(y_test_pca, y_pred_log_reg, average='binary')

# Print and save accuracy, precision, recall, and F1 score
print(f"Logistic Regression Model Accuracy on PCA-Transformed Test Data: {accuracy_log_reg * 100:.2f}%")
print(f"Precision: {precision_log_reg:.2f}")
print(f"Recall: {recall_log_reg:.2f}")
print(f"F1 Score: {f1_log_reg:.2f}")

# Confusion Matrix
conf_matrix_log_reg = confusion_matrix(y_test_pca, y_pred_log_reg)
disp_log_reg = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_log_reg, display_labels=['No ASD', 'ASD'])
disp_log_reg.plot(cmap='Blues')
plt.title('Confusion Matrix - Logistic Regression Model on PCA Data')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# Create a synthetic 2D dataset for classification
X, y = make_classification(n_samples=100, n_features=2, n_classes=2, n_clusters_per_class=1, n_informative=2, n_redundant=0, random_state=42)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the KNN model
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Plotting decision boundary
h = 0.02  # Mesh step size
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Predicting classification for each point in the mesh
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plotting
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o', s=100, cmap=plt.cm.coolwarm)
plt.title("KNN Decision Boundary with k=5")
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.decomposition import PCA

# Reduce data to 2D using PCA
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Train SVM classifier on 2D PCA-transformed data
svm_model_2d = SVC(kernel='linear', probability=True, random_state=42)
svm_model_2d.fit(X_train_pca, y_train)

# Create a mesh grid for decision boundary plotting
h = 0.02
x_min, x_max = X_train_pca[:, 0].min() - 1, X_train_pca[:, 0].max() + 1
y_min, y_max = X_train_pca[:, 1].min() - 1, X_train_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Plot decision boundary
Z = svm_model_2d.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, edgecolors='k', s=100, cmap=plt.cm.coolwarm)
plt.title("SVM Decision Boundary in 2D (PCA)")
plt.xlabel('PCA Feature 1')
plt.ylabel('PCA Feature 2')
plt.show()
```
